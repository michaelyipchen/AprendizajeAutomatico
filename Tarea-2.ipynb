#### Seleccion y Manipulacion

Elimina columnas Pregnancies y Glucose

```python
df.drop(['Pregnancies', 'Glucose'], axis=1)
```

Definicion de subset subdf2 para ejemplos siguientes.
Slice de filas 9 al 20 con columnas 3 y 4

```python
subdf2 = df.iloc[9:20, [3, 4]]
subdf2
```

Eliminar filas repetidas

```python
subdf2.drop_duplicates()
```

Truncar dataframe despues o antes de un indice

```python
df.truncate(after=5)
```

Definicion de subset subdf3 para ejemplos siguientes.
Se reemplazan por NaN en elementos del subset donde se tienen 0

```python
subdf3 = subdf2.mask(subdf2 == 0)
subdf3
```

#### Manejo de Valores Faltantes

Eliminar filas con valores NaN

```python
subdf3.dropna()
```

Rellenar o Reemplazar valores NaN por 0, u cualquier otro valor definido. Tomando como base el subset **subdf3**

```python
subdf4 = subdf3.fillna(0)
subdf4
```

Reemplazo de valores u condiciones por nuevo valor sustituto

```python
subdf4.replace(0, value=0.0005)
```

Definicion de subset subdf5 para ejemplos siguientes. Tomando como base el subset **subdf3**
Truncar valores despues de la fila 13

```python
subdf5 = subdf3.truncate(before=13)
subdf5
```

Interpolacion por defecto el cual es el promedio del valor anterior y posterior a la fila el cual contiene un NaN

```python
subdf5.interpolate()
```

Interpolacion por metodo nearest que es el valor anterior.

```python
subdf5.interpolate(method='nearest')
```

#### Ordenamiento y cambio de Forma

Ordenar valores segun la columna Insulin, y ubicar los valores NaN al final.

```python
subdf5.sort_values(by='Insulin', na_position='last')
```

Dataframe transpuesto

```python
subdf5.T
```
