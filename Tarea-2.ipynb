### Indices e Iteraciones

```python
# seleccion por nombre columnas
df.loc[:, ['Glucose', 'Outcome']]
```

```python
# seleccion por posicion de columna
df.iloc[:, [3, 8]]
```

```python
# seleccion por posicion de filas y columnas especificas
df.iloc[[5, 4], [3, 8]]
```

```python
# verifica si el dataframe contiene los valores dentro del parametro de la lista
df.isin([35])
```

```python
subdf = df.iloc[[5, 4], [3, 8]]
subdf
```

```python
# reemplaza por el parametro other dependiendo de la condicion dada, cuando este es False
subdf.where(subdf <= 0, other='hello')
```

```python
# similar a where, pero cuando la condicion evaluada es True
subdf.mask(subdf > 0, other='hello')
```
