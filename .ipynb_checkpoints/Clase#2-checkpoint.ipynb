# Algunas pruebas de pre-procesado

Uso de la media o la mediana para obtener un valor faltante

```python
from statistics import mean, median
```

```python
data = [17, 80, 22, 16, 20]
```

```python
mn = mean(data)
```

```python
print(data) # outlier es 80
```

```python
print(mn) 
```

```python
data2 = [17, 22, 16, 20]
```

```python
mn2 = mean(data2)
print(mn2)
```

```python
md = median(data)
print(md)
```

## Caso de pre-procesado con Diabetes

```python
import pandas as pd
from pandas import read_csv
import numpy as np
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/fmezacr/DataScienceTEC/master/Clase2/pima-indians-diabetes.csv')
```

```python
dataset.shape
```

```python
print(dataset.head(20))
```

```python
print(dataset.tail(20))
```

```python
print(dataset.describe())
```

```python
print((dataset == 0).astype(int).sum()) # cantidad de 0's por columna
```

```python
dataset[['6', '148', '72', '35', '0', '33.6']] = dataset[['6', '148', '72', '35', '0', '33.6']].replace(0, np.NaN)
print((dataset == 0).astype(int).sum())
print(dataset.tail(20))
```

```python
print((dataset.isnull()).sum())
```

```python
dataset.fillna(dataset.median(), inplace=True)
```

```python
print(dataset.tail(20))
```

```python
new_dataset = dataset.rename({'6': 'preg'}, axis = 'columns')
```

# Caso Titanic

```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/fmezacr/DataScienceTEC/master/Clase2/titanic.csv')
```

```python
cols_to_use = ['Pclass', 'Sex', 'Age', 'Fare', 'Survived']
```

```python
data = data[cols_to_use]
```

```python
data.isnull().values.any()
```

```python
data.shape
```

```python
data = data.dropna()
```

```python
data.shape # para ver si el crtiero limpia un poco los datos de forma decente
```

```python
data['Sex'] = data['Sex'].replace({'male': 0, 'female': 1})
```

```python
data.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = data[['Sex', 'Age', 'Fare']]
labels = data['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, labels, random_state=13) # 75/25 por defecto
```
