## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
```

## Carga de archivo CSV
Se carga el archivo csv hacia un dataframe *daily_min_temperatures* para ser usado globalmente


```python
daily_min_temperatures = pd.read_csv('daily-min-temperatures.csv')
#print(daily_min_temperatures)
```

## Definicion de funciones manuales
- Normalizar
- Estandarizar

```python
def normalizar_manual(x):
    return (x - min(x))/(max(x) - min(x))

def estandarizar_manual(x):
    return (x - x.mean())/(x.std())
```

## Pruebas de funciones manuales

```python
temp = daily_min_temperatures.iloc[:, 1:2].values

#print(normalizar_manual(temp))
#print(estandarizar_manual(temp))

daily_min_temperatures_normalized = daily_min_temperatures
daily_min_temperatures_standarized = daily_min_temperatures

daily_min_temperatures_normalized.iloc[:, 1:2] = normalizar_manual(temp)
print("Normalizacion Manual\n", daily_min_temperatures_normalized)

daily_min_temperatures_standarized.iloc[:, 1:2] = estandarizar_manual(temp)
print("\nEstandarizacion Manual\n", daily_min_temperatures_standarized)
```

## Definicion de funciones automaticas o por bibliotecas scikit learn
- Normalizar
- Estandarizar

```python
def normalizar_automatico(x):
    # definicion de rango de escala
    scaler = MinMaxScaler(feature_range=(0, 1))
    
    # aplicar y  transformar con scaler
    scaler = scaler.fit(x)
    normalized_x = scaler.transform(x)
    return normalized_x

def estandarizar_automatico(x):
    # aplicar y  transformar con scaler de normalizacion standard
    scaler = preprocessing.StandardScaler()
    scaler = scaler.fit(x)
    standarized_x = scaler.transform(x)
    return standarized_x
```

## Pruebas de funciones automaticas o por bibliotecas scikit learn

```python
temp = daily_min_temperatures.iloc[:, 1:2].values

#print(normalizar_automatico(temp))
#print(estandarizar_automatico(temp))

daily_min_temperatures_normalized = daily_min_temperatures
daily_min_temperatures_standarized = daily_min_temperatures

daily_min_temperatures_normalized.iloc[:, 1:2] = normalizar_automatico(temp)
print("Normalizacion Automatica\n", daily_min_temperatures_normalized)

daily_min_temperatures_standarized.iloc[:, 1:2] = estandarizar_automatico(temp)
print("\nEstandarizacion Automatica\n", daily_min_temperatures_standarized)
```
